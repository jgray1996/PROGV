```python
def genbank_parser(record):
    annotations = record.annotations
    # Accession number
    accession_number = annotations["accessions"]
    # Size of Genome in basepairs
    genome_size = len(record.features[0].location)
    # Bioproject, Biosample and Assembly
    bioproject, biosample, assembly = record.dbxrefs
    # First publication and year of publication
    references = annotations["references"][0]
    first_publication = references.pubmed_id
    first_publication_year = references.journal[-5:-1]
    # Current genbank version
    current_version_gb = annotations["date"]

    # Fylogeny and species
    organism = annotations["organism"]
    taxonomy = annotations["taxonomy"]
    return accession_number, genome_size, bioproject.split(":")[1], biosample.split(":")[1], assembly.split(":")[1], first_publication, first_publication_year, current_version_gb, organism, taxonomy


def calc_max
```
